## UTILIZZO DISTRIBUZIONE UNIFORME

```python
#controlla statistiche per distribuzione uniforme cambiando per ogni esecuzione i valori della distribuzione
#n = numero di volte che si vuole ripetere l'operazione
#S = seller
#B = buyer
#results = lista risultati finali per ogni esecuzione

def create_uniform_distribution(num_round,num_cifre):   
    S_U = np.around(np.random.uniform(0,1, num_round), num_cifre)
    B_U = np.around(np.random.uniform(0,1, num_round), num_cifre)
    
    graph_values = [0, 1, 0, 1]
    
    return S_U, B_U, graph_values
```

```python
#controlla statistiche per distribuzione uniforme cambiando per ogni esecuzione i valori della distribuzione
#creiamo distribuzioni uniformi che non hanno valore in comune
#n = numero di volte che si vuole ripetere l'operazione
#S = seller
#B = buyer
#results = lista risultati finali per ogni esecuzione
#spaced_value in [0,1]= valore di spazio tra estremi fine-inizio delle distribuzioni

def create_uniform_distribution_spaced(num_round,num_cifre):
    spaced_value_s = np.around(np.random.rand(),1)
    spaced_value_b = np.around(np.random.uniform(spaced_value_s,1)) #in questo modo spaced_value_b > spaced_value_s
    
    S_U = np.around(np.random.uniform(0,spaced_value_s, num_round), num_cifre)
    B_U = np.around(np.random.uniform(spaced_value_b,1, num_round), num_cifre)
    
    graph_values = [0, 1, 0, 1]
    
    return S_U, B_U, graph_values
```

```python
def draw_uniform_distribution(a_s,b_s,a_b,b_b,num):
    fig, axs = plt.subplots(1,2,figsize = (15,5))
    B = st.uniform(a_b,b_b)
    
    fig.suptitle("Distribuzione Seller e Buyer", fontsize = 20)
    s = np.linspace(st.uniform.ppf(0.01,a_s,b_s),st.uniform.ppf(0.99,a_s, b_s), num)
    b = np.linspace(B.ppf(0.01),B.ppf(0.99), num)
    
    axs[0].set_title("Distribuzione Seller")
    axs[0].plot(s,st.uniform.pdf(s,a_s,b_s))
    
    axs[1].set_title("Distribuzione Buyer")
    axs[1].plot(b,B.pdf(b))
```

```python
import random
def create_beta_distribution(num_round, num_cifre):

    #alpha_seller = float(input("inserire valore alpha seller: "))
    #beta_seller = float(input("inserire valore beta seller: "))
    alpha_seller = random.uniform(0, 1)
    beta_seller = random.uniform(0, 1)
    print("valore alpha seller: ",alpha_seller)
    print("valore beta seller: ",beta_seller)

    #alpha_buyer = float(input("inserire valore alpha buyer: "))
    #beta_buyer = float(input("inserire valore beta buyer: "))
    alpha_buyer = random.uniform(0, 1)
    beta_buyer = random.uniform(0, 1)
    print("valore alpha buyer: ",alpha_buyer)
    print("valore beta buyer: ",beta_buyer)

        
    S_D = np.around(np.random.beta(alpha_seller, beta_seller, num_round), num_cifre)
    B_D = np.around(np.random.beta(alpha_buyer, beta_buyer, num_round), num_cifre)
        
    print("alpha seller: ", alpha_seller, "\tbeta seller: ",
        beta_seller, "\nalpha buyer: ", alpha_buyer, "\tbeta buyer: ", beta_buyer)
    
    graph_values = [alpha_seller, beta_seller, alpha_buyer, beta_buyer]
        
    return S_D, B_D, graph_values
```

```python
#disegna la funzione di densità per il seller e il buyer

def draw_pdf_beta(alpha_seller, beta_seller, alpha_buyer, beta_buyer, num):
    
    fig, axs = plt.subplots(1,2,figsize = (15,5))

    fig.suptitle("Distribuzione Seller e Buyer", fontsize = 20)
    s = np.linspace(st.beta.ppf(0.01, alpha_seller, beta_seller),st.beta.ppf(0.99, alpha_seller, beta_seller), num)
    b = np.linspace(st.beta.ppf(0.01, alpha_buyer, beta_buyer),st.beta.ppf(0.99, alpha_buyer, beta_buyer), num)
    
    limit = max([alpha_seller, beta_seller, alpha_buyer, beta_buyer])#limite y per entrambi
    
    axs[0].set_title("Distribuzione Seller")
    axs[0].plot(s, st.beta.pdf(s, alpha_seller, beta_seller), 'r-')
    axs[0].set_ylim(0, limit)
    axs[0].set_xlim([0,1])
    
    axs[1].set_title("Distribuzione Buyer")
    axs[1].plot(b, st.beta.pdf(b, alpha_buyer, beta_buyer), 'g-')
    axs[1].set_ylim(0, limit)
    axs[1].set_xlim([0,1])
```

### ATTRAVERSO METODO DELL'INVERSIONE CON CONCENTRAZIONE SUI PUNTI $x_1,..,x_n$ $\subset$ [0,1]

Ma stavolta giusto. Cioè?

Definire n bins nell'intervallo tra 0 e 1 e inserire degli elementi compresi tra questi intervalli.
Il peso di ciascun intervallo è il numero di elementi in esso.

```python
#controlla statistiche per distribuzione uniforme cambiando per ogni esecuzione i valori della distribuzione
#n = numero di volte che si vuole ripetere l'operazione
#S = seller
#B = buyer
#results = lista risultati finali per ogni esecuzione

def create_discretaRND_distribution(num_round,num_cifre):   
    
    S_D_tmp, _, _ = generate_random_weighted_distribution(num_round, num_cifre)
    B_D_tmp, _,_ = generate_random_weighted_distribution(num_round, num_cifre)
    
    S_D = np.concatenate(S_D_tmp).tolist()
    B_D = np.concatenate(B_D_tmp).tolist() #lista di valori per la distribuzione buyer e seller pesate    
    
    graph_values = [0, 1, 0, 1]
    
    return S_D, B_D, graph_values
```

```python
#controlla statistiche per distribuzione uniforme cambiando per ogni esecuzione i valori della distribuzione
#n = numero di volte che si vuole ripetere l'operazione
#S = seller
#B = buyer
#results = lista risultati finali per ogni esecuzione

def create_discretaUSR_distribution(num_round,num_cifre):   
    S_D_tmp, _, _ = generate_user_weighted_distribution(num_round, num_cifre)
    B_D_tmp, _,_ = generate_user_weighted_distribution(num_round, num_cifre)
    
    S_D = np.concatenate(S_D_tmp).tolist()
    B_D = np.concatenate(B_D_tmp).tolist() #lista di valori per la distribuzione buyer e seller pesate    
    
    graph_values = [0, 1, 0, 1]
    
    return S_U, B_U, graph_values
```

```python
#l'utente definisce gli intervalli in cui saranno presenti gli elementi della distribuzione discreta
#restituisce dizionario contenente {k = l_b ; }
def define_user_bins():
    interval_list = []
    num_bins = int(input("Inserire numero di intervalli: "))
    for i in range(0,num_bins):
        print("\nDefinire l'i-esimo intervallo: ")
        
        if(i == 0):lower_bound = 0
        else: lower_bound = interval_list[-1]
        
        if(i == num_bins-1): upper_bound = 1
        else: upper_bound = float(input("Upper bound: "))
        
        if(lower_bound > upper_bound): 
            print("Errore, valori sbagliati -> lower_bound > upper_bound")
            while(lower_bound > upper_bound): 
                lower_bound,upper_bound = int(input("Lower bound: ")),int(input("Upper bound: "))
        print("Lower bound: ", lower_bound, "\tUpper bound: ", upper_bound)
        
        interval_list.append(upper_bound)
        
    return interval_list
```

```python
#definiamo randomicamente gli intervalli in cui saranno presenti gli elementi della distribuzione discreta
#restituisce dizionario contenente {k = l_b ; }
def define_random_bins(num_cifre, num_max_intervals):
    interval_list = []
    num_bins = np.random.randint(1,11)  #11 è un numero arbitrario in modo tale che il numero di intervalli sia tra 1 e 10 
                                        #-> ad esempio [0-0.1 ; 0.1-0.2 ; 0.2-0.3 ;...;0.9-1.0]
    for i in range(0,num_bins):
        print("\nDefinire l'i-esimo intervallo: ")
        
        if(i == 0):lower_bound = 0
        else: lower_bound = interval_list[-1]
        
        if(i == num_bins-1): upper_bound = 1
        else: upper_bound = np.round(np.random.random(), num_cifre)
            
        if(lower_bound > upper_bound): 
            while(lower_bound > upper_bound):
                lower_bound,upper_bound = interval_list[-1], np.round(np.random.random(), num_cifre)
        print("Lower bound: ", lower_bound, "\tUpper bound: ", upper_bound)
        
        interval_list.append(upper_bound)
        
    return interval_list
```

```python
#genera la distribuzione e i pesi ad essa associati secondo le specifiche dell'utente
def generate_user_weighted_distribution(num_round, num_cifre):
    interval_list = define_user_bins()
    count = num_round
    num_of_els_in_intervals = [] #lista che contiene il peso di ciascun intervallo
    
    for i in range(0, len(interval_list)):        
        if(i == len(interval_list)-1): num_of_els = count #nel caso in cui siamo all'ultimo intervallo, il suo peso è pari a 
                                                        #num_round - round_assegnati_fino_ad_adesso
        
        else: 
            num_of_els = int(input("Numero di elementi per l'i-esimo intervallo")) #numero di elementi per l'i-esimo intervallo
            while(num_of_els > count): num_of_els = int(input("Errore, inserire numero di elementi per l'i-esimo intervallo"))
            
        count -= num_of_els #numero di elementi rimasti da cui scegliere
        
        #nel caso in cui abbiamo finito il numero di elementi da inserire ma abbiamo ancora altri intervalli
        if((count == 0) & (i != len(interval_list)-1)): num_of_els = 0 #gli altri intervalli avranno tutti peso 0
            
        num_of_els_in_intervals.append(num_of_els) #inserisci peso
        
    count_bin = 0
    values_of_num_of_els = []
    for num_of_els in num_of_els_in_intervals:
        #definiamo lower bound
        if(count_bin == 0): lower_bound = 0  #se siamo al primo intervallo allora lower bound è sempre 0 ma non è detto!!
        else: lower_bound = interval_list[count_bin-1]
        
        #definiamo upper bound
        if(count_bin == len(num_of_els_in_intervals) - 1): upper_bound = 1 #se siamo all'ultimo intervallo allora upper bound è sempre 1 ma non è detto!!
        else: upper_bound = interval_list[count_bin]
            
        values = np.round(np.random.uniform(lower_bound,upper_bound, num_of_els), num_cifre)
    
        count_bin+=1
        
        values_of_num_of_els.append(values)
    return values_of_num_of_els, num_of_els_in_intervals ,interval_list
```

```python
#genera randomicamente la distribuzione e i pesi ad essa associati
def generate_random_weighted_distribution(num_round, num_cifre):
    num_max_intervals = np.random.rand(1,20)
    interval_list = define_random_bins(num_cifre,num_max_intervals)
    count = num_round
    num_of_els_in_intervals = [] #lista che contiene il peso di ciascun intervallo
    
    for i in range(0, len(interval_list)):        
        if(i == len(interval_list)-1): num_of_els = count #nel caso in cui siamo all'ultimo intervallo, il suo peso è pari a 
                                                        #num_round - round_assegnati_fino_ad_adesso
        else:
            num_of_els = np.random.randint(0, count) #numero di elementi per l'i-esimo intervallo
            while(num_of_els > count ): num_of_els = np.random.randint(0, count)
            
        count -= num_of_els #numero di elementi rimasti da cui scegliere
        
        #nel caso in cui abbiamo finito il numero di elementi da inserire ma abbiamo ancora altri intervalli
        if((count == 0) & (i != len(interval_list)-1)): num_of_els = 0 #gli altri intervalli avranno tutti peso 0
            
        num_of_els_in_intervals.append(num_of_els) #inserisci peso
    
    count_bin = 0
    values_of_num_of_els = []
    for num_of_els in num_of_els_in_intervals:
        #definiamo lower bound
        if(count_bin == 0): lower_bound = 0  #se siamo al primo intervallo allora lower bound è sempre 0 ma non è detto!!
        else: lower_bound = interval_list[count_bin-1]
        
        #definiamo upper bound
        if(count_bin == len(num_of_els_in_intervals) - 1): upper_bound = 1 #se siamo all'ultimo intervallo allora upper bound è sempre 1 ma non è detto!!
        else: upper_bound = interval_list[count_bin]
            
        values = np.round(np.random.uniform(lower_bound,upper_bound, num_of_els), num_cifre)
    
        count_bin+=1
        
        values_of_num_of_els.append(values)
        
    return values_of_num_of_els, num_of_els_in_intervals ,interval_list
```

## Distribuzione D± $\epsilon$

Distribuzione del buyer: $f_{S, \pm\epsilon} = 2(1\pm\epsilon) * I[0, \frac{1}{4}] + 2(1\mp\epsilon) * I[\frac{1}{2}, \frac{3}{4}]$

Distribuzione del seller: $f_{B} = 2 * (I[\frac{1}{4}, \frac{1}{2}] \cup I[\frac{3}{4}, 1])$

```python
# funzione densità seller
def f_S(e, x):
    if x < 0 or x > 3/4:
        return 0
    if x <= 1/4:
        return 2*(1+e)
    elif (x <= 3/4) and (x >= 0.5):
        return 2*(1-e)
    else:
        return 0
```

```python
# funzione densità buyer
def f_B(x):
    if x < 1/4 or x > 1:
        return 0
    elif x <= 1/2 or x >= 3/4:
        return 2
    else:
        return 0
```

```python
import numpy as np
import matplotlib.pyplot as plt


def FBP_epsilon_density_functions(epsilon):
    def distribution_S(epsilon):
        def f_S_positive(x):
            return 2 * (1 + epsilon) if 0 <= x <= 1/4 else 0
        
        def f_S_negative(x):
            return 2 * (1 - epsilon) if 1/2 <= x <= 3/4 else 0
        
        return np.vectorize(lambda x: f_S_positive(x) + f_S_negative(x))
    
    def distribution_B():
        def f_B(x):
            return 2 if (1/4 <= x < 1/2) or (3/4 <= x <= 1) else 0
        
        return np.vectorize(f_B)
    
    x = np.linspace(0, 1, 1000)
    y_S = distribution_S(epsilon)(x)
    y_B = distribution_B()(x)
    
    plt.fill_between(x, y_S, color="green", alpha=0.5, label="Distribuzione S")
    plt.fill_between(x, y_B, color="yellow", alpha=0.5, label="Distribuzione B")
    plt.xlabel("p")
    plt.ylabel("PDF")
    plt.title("Funzione di densità Seller (S) e Buyer (B)")
    plt.legend()
    plt.grid()
    plt.show()
```
